```python
rating_movie_animation = pd.read_csv("/content/drive/MyDrive/data.csv")
top_movie=rating_movie_animation.sort_values(["numVotes","averageRating"],ascending=False)
top_movie=top_movie[0:5]
top_movie.head()
```

```python
#classement films d'animation
#la largeur et la hauteur de la figure

sns.set_style("dark")#Je définie un thème

palette=sns.color_palette("Paired")#Je choisi une palette de couleur
sns.set_palette(palette)

plt.figure(figsize=(18,8))#Taille du graphique
plt.subplot(111)#Position du graphique
sns.barplot(data=top_movie , x="originalTitle", y="averageRating", hue="numVotes")#Création du barplot
plt.title("Meilleurs films d'animation en fonction du nombres de vote", fontsize=30)#Titre
plt.xticks(rotation=45, fontsize=13)#rotation des tiquet de l"axe x avec taille
plt.xlabel("Titre", fontsize=20)#Titre de l'axe X
plt.ylabel("Note du public", fontsize=20)#Titre de l'axe Y
plt.legend(title = 'Nombres de votes',fontsize=15)#MChanngement du titre de la legende



```
