## Initialiser fonction reco_to_list

```python
def reco_to_list(reco_):
  len(reco_[0])
  liste_id = []
  for n in range(len(reco_[0])):
    liste_id.append(reco_[0][n])
  return liste_id
```

## Variables contenant les résultats de recommandation (x nombre de modèles)

```python
#Appel des voisins basés sur GENRE
distance_genre, id_proches_genre = model_genre.kneighbors(X_genre.loc[df_final["originalTitle"]== film_user],n_neighbors=10)

#Appel des voisins basés sur DIRECTOR+DATE
distance_director_date, id_proches_director_date = model_director_date.kneighbors(X_director_date.loc[df_final["originalTitle"]== film_user],n_neighbors=10)


#Appel des voisins basés sur GENRE+DATE
distance_genre_date, id_proches_genre_date = model_genre_date.kneighbors(X_genre_date.loc[df_final["originalTitle"]== film_user],n_neighbors=10)


#Appel des voisins TOPRATING + GENRE
distance_top_genre, id_proches_top_genre = model_genre.kneighbors(X_genre.loc[df_final["originalTitle"]== film_user],n_neighbors=100)

#Appel des voisins TOPRATING + DIRECTOR
distance_top_director, id_proches_top_director = model_director.kneighbors(X_director.loc[df_final["originalTitle"]== film_user],n_neighbors=100)

#Appel des voisis TOPRATING+GENRE+DATE
distance_top_genre_date, id_proches_top_genre_date = model_genre_date.kneighbors(X_genre_date.loc[df_final["originalTitle"]== film_user],n_neighbors=100)
```
